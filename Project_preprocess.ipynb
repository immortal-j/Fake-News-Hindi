```python
import json
import pandas as pd
import re
with open('fake_news.json') as f:
  fake_news = json.load(f)
  #json to dataframe
  fake_news = pd.json_normalize(fake_news,'fake_news')
  f.close()

with open('true_news.json') as f:
  true_news = json.load(f)
  #json to dataframe
  true_news = pd.json_normalize(true_news,'true_news')
  f.close()
```

```python
true_news = true_news.rename(columns={'short_desc': 'short_description'})
```

```python
true_news=true_news.drop(['name','url'],axis=1)
fake_news=fake_news.drop(['url'],axis=1)
```

```python

true_news=true_news.drop(['full_title','long_description',],axis=1)
fake_news=fake_news.drop(['full_title', 'long_description'],axis=1)

```

```python
fake_news=fake_news.dropna()
true_news=true_news.dropna()
true_news
```

```python
#Removing Seperators

def remove_sep( news ):
 for i in list(true_news):
  news[i]=news[i].str.replace('|', '')
  news[i]=news[i].str.replace('?', '')
  news[i]=news[i].str.replace(':', '')
  news[i]=news[i].str.replace(';', '')
  news[i]=news[i].str.replace("'", '')
  news[i]=news[i].str.replace('"', '')
  news[i]=news[i].str.replace(',', '')
  news[i]=news[i].str.replace('.', '')
  news[i]=news[i].str.replace('(', '')
  news[i]=news[i].str.replace(')', '')
  news[i]=news[i].str.replace('\n', '')
  news[i]=news[i].str.replace('&', '')
  news[i]=news[i].str.replace('।', '')
  

remove_sep(true_news)
remove_sep(fake_news)

```

```python
#Stemming
suffixes = {
    # 1: ["ो", "े", "ू", "ु", "ी", "ि", "ा"],
    1: ["कर", "ाओ", "िए", "ाई", "ाए", "ने", "नी", "ना", "ते", "ीं", "ती", "ता", "ाँ", "ां", "ों", "ें"],
    2: ["ाकर", "ाइए", "ाईं", "ाया", "ेगी", "ेगा", "ोगी", "ोगे", "ाने", "ाना", "ाते", "ाती", "ाता", "तीं", "ाओं", "ाएं", "ुओं", "ुएं", "ुआं"],
    3: ["ाएगी", "ाएगा", "ाओगी", "ाओगे", "एंगी", "ेंगी", "एंगे", "ेंगे", "ूंगी", "ूंगा", "ातीं", "नाओं", "नाएं", "ताओं", "ताएं", "ियाँ", "ियों", "ियां"],
    4: ["ाएंगी", "ाएंगे", "ाऊंगी", "ाऊंगा", "ाइयाँ", "ाइयों", "ाइयां"],
}

def hi_stem(word):
    for L in  4, 3, 2, 1:
        if len(word) > L + 1:
            for suf in suffixes[L]:
                if word.endswith(suf):
                    return word[:-L]
    return word
```

```python
id1=list(fake_news.index)
id2=list(true_news.index)
```

```python
str_temp=""
count=0
for i in list(fake_news):
  count=0
  for j in list(fake_news[i]):
    for words in j.split():
      str_temp+=hi_stem(words)
      str_temp+=" "
    fake_news.loc[id1[count],i]=str_temp
    str_temp=""
    count+=1

    
```

```python
str_temp=""
count=0
for i in list(true_news):
  count=0
  for j in list(true_news[i]):
    for words in j.split():
      str_temp+=hi_stem(words)
      str_temp+=" "
    true_news.loc[id2[count],i]=str_temp
    str_temp=""
    count+=1
```
